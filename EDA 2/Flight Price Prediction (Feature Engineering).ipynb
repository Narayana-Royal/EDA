```python
#step 1 : Importing the necessary libraries and reading datasets using pandas
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df_train = pd.read_excel("Data_Train.xlsx")
```

```python
df_test = pd.read_excel("Test_set.xlsx")
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
#step 2 : Combining two datasets as single dataframe
```

```python
df = pd.concat([df_train,df_test])
df.tail()
```

```python
#step 3 : Applying Feature Engineering on all columns as all are required to transform

#step 3.1 - STARTING WITH COLUMN: Date_of_Journey
```

```python
df["Date"]=df["Date_of_Journey"].str.split("/").str[0]
df["Month"]=df["Date_of_Journey"].str.split("/").str[1]
df["Year"]=df["Date_of_Journey"].str.split("/").str[2]
```

```python
df.drop(["Date_of_Journey"],axis=1,inplace=True)
```

```python
df["Date"]=df["Date"].astype(int)
df["Month"]=df["Month"].astype(int)
df["Year"]=df["Year"].astype(int)
```

```python
df.head(3)
```

```python
#step 3.2 - Next taking COLUMN: Arrival Time
```

```python
df["Arrival_Hour"]=df["Arrival_Time"].str.split(" ").str[0].str.split(":").str[0]
df["Arrival_Min"]=df["Arrival_Time"].str.split(" ").str[0].str.split(":").str[1]
```

```python
df["Arrival_Hour"]=df["Arrival_Hour"].astype(int)
df["Arrival_Min"]=df["Arrival_Min"].astype(int)
```

```python
df.head(3)
```

```python
df.drop(["Arrival_Time"],axis=1,inplace=True)
```

```python
#step 3.2 - Next taking COLUMN: Dept Time
```

```python
df["Dep_Hour"]=df["Dep_Time"].str.split(":").str[0]
```

```python
df["Dep_Hour"]=df["Dep_Hour"].astype(int)
```

```python
df.head(3)
```

```python
df.drop(["Dep_Time"],axis=1,inplace=True)
```

```python
#step 3.2 - Next taking COLUMN: Duration
```

```python
df["Duration_Hour"]=df["Duration"].str.split("h").str[0]
```

```python
df[df["Duration"]=="5m"]
```

```python
df.drop([6474,2660],inplace=True)
```

```python
df["Duration_Hour"]=df["Duration_Hour"].astype(int)
```

```python
df.drop(["Duration"],axis=1,inplace=True)
```

```python
df.head(3)
```

```python
#step 3.3 - Next taking COLUMN: Total_Stops
```

```python
df["Total_Stops"].unique()
```

```python
df["Total_Stops"]=df["Total_Stops"].map({'non-stop':0,'2 stops':2,'1 stop':1,'3 stops':3,'4 stops':4,"nan":1})
```

```python
df.head()
```

```python
#step 3.4 - Next taking COLUMN: Route
```

```python
df.drop(["Route"],axis=1,inplace=True)
```

```python
df.head()
```

```python
#step 3.5 - Next taking all the remaining COLUMNS: Airline, Source, Destination, Additional_Info
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

df["Airline"] = labelencoder.fit_transform(df["Airline"])
df["Source"] = labelencoder.fit_transform(df["Source"])
df["Destination"] = labelencoder.fit_transform(df["Destination"])
df["Additional_Info"] = labelencoder.fit_transform(df["Additional_Info"])
```

```python
df.head(3)
```
